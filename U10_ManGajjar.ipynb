```python
# anti clockwise
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([15,25,25,35])
mylabesl = ["Apple","Banana","chreeris","Dates"]
plt.pie(x,labels=mylabesl,startangle=90)
plt.show()
```

```python
# clockwise
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([15,25,25,35])
mylabesl = ["Apple","Banana","chreeris","Dates"]
plt.pie(x,labels=mylabesl,startangle=90,counterclock=False)
plt.show()
```

```python
# explode
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([15,25,25,35])
mylabesl = ["Apple","Pizzo","chreeris","Dates"]
myexplode=[0,0.5,0,0]
plt.pie(x,labels=mylabesl,startangle=90,explode=myexplode)
plt.show()
```

```python
# shadow
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([15,25,25,35])
mylabesl = ["Apple","Pizzo","chreeris","Dates"]
myexplode=[0,0.5,0,0]
plt.pie(x,labels=mylabesl,startangle=90,explode=myexplode,counterclock=False,shadow=True)
plt.show()
```

```python
# legend
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([15,25,25,35])
mylabesl = ["Apple","Pizzo","chreeris","Dates"]
plt.pie(x,labels=mylabesl,startangle=90,counterclock=False)
plt.legend()
plt.show()
```

```python
# Size bigger
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([15,25,25,35])
plt.figure(figsize=(8,8))
mylabesl = ["Apple","Pizzo","chreeris","Dates"]
plt.pie(x,labels=mylabesl,autopct="%1.2f%%")
plt.legend(title="Four Fruits",loc="right")
plt.show()
```

```python
# pie chart for popullary data 
# creat and expload view of py and sho the % each py pro.


import matplotlib.pyplot as plt 
import numpy as np

plt.figure(figsize=(8,10),facecolor="#C4D9FF")
x = np.array([29,19,8,7,6,5,3])
mylabesl = ["Python","Java","Javascript","C#","PHP","C,C++","R"]
myexplode=[0.2,0,0,0,0,0,0]
plt.pie(x,labels=mylabesl,autopct="%1.2f%%",explode=myexplode)
plt.legend(title="Language",loc="lower right")
plt.show()
```
